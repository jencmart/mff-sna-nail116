## 1. getWikipediaNetwork(ego,depth)

```python
import warnings
import networkx as nx
import wikipedia
import matplotlib.pyplot as plt

wikipedia.set_lang("en")
warnings.filterwarnings("ignore", category=UserWarning, module='wikipedia')


def get_page(page, remaining_depth):
    wiki = None
    try:
        wiki = wikipedia.page(page)
    except wikipedia.DisambiguationError:
        pass # print("[AMBIGUOUS]{} ".format(page))
    except wikipedia.PageError:
        try:
            wiki = wikipedia.page(page,auto_suggest=False)
        except: 
            pass # print("[NO MATCH]{} ".format(page))
    except:
        pass # print("[CAN'T LOAD]{} ".format(page))
    return wiki
    
    
def get_recursive(G, curr_node_link, remaining_depth):            
    if "(identifier)" in curr_node_link or "(Identifier)" in curr_node_link:
        return
    wiki = get_page(curr_node_link, remaining_depth)
    if wiki is None or "(identifier)" in wiki.title or "(Identifier)" in wiki.title:
        return

    G.add_node(curr_node_link)
    if remaining_depth == 0:
        return
  
    for idx, link in enumerate(wiki.links):
        if idx%20==0 or remaining_depth==2:
            for i in range(2-remaining_depth):
                print("\t", end="")
            print("{}/{} ---> {}".format(idx, len(wiki.links), link ))
        G.add_node(link)
        G.add_edge(curr_node_link, link)
        get_recursive(G, link, remaining_depth-1)
        
    
def getWikipediaNetwork(ego,depth=2):
    # oriented network of type `nx.DiGraph()`  around page with title `name` of depth `depth`. 
    G=nx.DiGraph()
    get_recursive(G,ego, 2)
    return G
```

## 2. getNetwork(name,depth=2,download=False)

```python
import os
import pandas as pd


def load_from_csv(name):
    lst = pd.read_csv(name+".csv").values.tolist()
    G = nx.DiGraph()
    G.add_edges_from(lst)
    return G


def save_to_csv(G, name):
    with open(name+".csv", 'w') as f:
        for e in G.edges:
            a = e[0].replace(",", "")
            b = e[1].replace(",", "")
            f.write("{},{}\n".format(a,b))
 
    
def getNetwork(name,depth=2,download=False):
    if not download and os.path.isfile(name + ".csv"):
        print("[{}]".format(name))
        print("-------------------------------------------------------------------------------------")
        return load_from_csv(name)
    else:
        print("[{}] ...downloading...".format(name))
        print("-------------------------------------------------------------------------------------")
        G = getWikipediaNetwork(name,depth=depth)
        save_to_csv(G, name)
        return G
```

## 3. analyzeEgo(G, ego, topNeighbours=20)

```python
def remove_nodes(G, ego):
    to_delete = []
    for node in G.nodes:
        if G.degree[node] <= 1:
            to_delete.append(node)
    if len(to_delete) != 0:
        G.remove_nodes_from(to_delete)

def analyzeEgo(G, ego, topNeighbours=20):
    
    # >>>>>>>>>>>>>>>>>>>> DELETE NODES  >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
    print("\toriginal:  |V|={}\t|E|={}".format(len(G.nodes), len(G.edges)))
    remove_nodes(G, ego)
    print("\ttruncated: |V|={}\t|E|={}".format(len(G.nodes), len(G.edges)))   
    # <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
    
    
    # >>>>>>>>>>>>>>>>>>>> CENTRALITY MEASURES >>>>>>>>>>>>>>>>>>>>>>>>>>>>
    print("-------------------------------------------------------------------------------------")
    outd = G.out_degree(ego)
    ind =  G.in_degree(ego) 
    max_edges = G.number_of_nodes() - 1
    assert max_edges > 0
    print("\tOut Degree Centrality:  Cd-({}) = {:.1f}\t(normalized = {:.4f})".format(ego, outd , outd/max_edges))
    print("\tIn  Degree Centrality:  Cd+({}) = {:.1f}\t(normalized = {:.4f})".format(ego, ind, ind/max_edges))
    assert outd + ind == G.degree(ego)
    
    bc = nx.betweenness_centrality(G, normalized=False)[ego]
    bc_norm = nx.betweenness_centrality(G, normalized=True)[ego]
    print("\tBetweenness Centrality: Cb({})  = {:.1f}\t(normalized = {:.4f})".format(ego, bc, bc_norm))
    
    cc_norm = nx.closeness_centrality(G.reverse(), u=ego) # reverse because we want to go from ego..
    cc = cc_norm*max_edges
    print("\tCloseness Centrality:   Cc({})  = {:.1f}\t(normalized = {:.4f})".format(ego, cc, cc_norm))
    print("-------------------------------------------------------------------------------------")
    # <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
    
    
    # >>>>>>>>>>>>>>>>>>>> TOP NEIGHBOURS >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>
    nodes = sorted(G.in_degree, key=lambda x: x[1], reverse=True)
    if ego in nodes:
        nodes.remove(ego) # topNeighbours without ego ...
    nodes = nodes[:topNeighbours]
    res = []
    for i in range (len(nodes)):
        res.append ( (nodes[i][0], nodes[i][1]) )
    # <<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<

    return res
```

## 4. nice plotting function

```python
def plot_ego_net(G,result, ego):
    sub = [n[0] for n in result] # get top-in-degree node names
    sub.append(ego) # and also add ego-node
    H = G.subgraph(sub)
    pos = nx.circular_layout(H)

    plt.figure(figsize=(12,10)) 

    # Nodes
    options = {"alpha": 0.2} # "node_size": 50, 
    nx.draw_networkx_nodes(H, pos, nodelist=[ego], node_color="r", **options)
    nx.draw_networkx_nodes(H, pos, nodelist=[n[0] for n in result], node_color="y", **options)

    
    # all edges ...
    nx.draw_networkx_edges(H, pos, width=1.0, alpha=0.1)
    
    # ego edges
    in_edges = H.in_edges(ego)
    out_edges = H.out_edges(ego) 
    nx.draw_networkx_edges(H,pos, edgelist=in_edges, width=1, alpha=0.4, edge_color="b")
    nx.draw_networkx_edges(H,pos, edgelist=out_edges, width=1, alpha=0.4, edge_color="r")

    # Labels (without ego)
    pos_no = pos.copy()
    del pos_no[ego]
    nx.draw_networkx_labels(H.subgraph([n[0] for n in result]), pos_no,font_size=12, 
                            bbox=dict(boxstyle="round,pad=0.1", fc="yellow", alpha=0.5))
    
    # Labels (ego only)
    ego_pos = {ego : pos[ego]}
    nx.draw_networkx_labels(H.subgraph(ego), ego_pos,font_size=13, 
                        bbox=dict(boxstyle="round,pad=0.2", fc="red", alpha=0.4))

```

# B. ANALYSIS

```python
egos = ["Network science", "Complex network", "Project Jupyter"]
top=20
```

## "Network science"

```python
ego = egos[0]
G = getNetwork(ego, depth=2, download=False)
result = analyzeEgo(G, ego, topNeighbours=top)

plot_ego_net(G,result, ego)
```

## "Complex network"

```python
ego = egos[1]
G = getNetwork(ego, depth=2, download=False)
result = analyzeEgo(G, ego, topNeighbours=top)

plot_ego_net(G,result, ego)
```

## "Project Jupyter"

```python
ego = egos[2]
G = getNetwork(ego, depth=2, download=False)
result = analyzeEgo(G, ego, topNeighbours=top)

plot_ego_net(G,result, ego)
```

##  Are the networks comparable?
* w.r.t to the size "Network science" is the smallest ego net, but latter two are very similar.

* "Network science" has a lot connections to neighbour edges (almost 40% of all nodes) compared to "Complex Network" and "Project Jupyer" (both under 5%)

* In degree is very small for all three egos.

* Betweenness Centrality is very similar in all three. So these egos probably not dot act as a bridges along the shortest path between their neigbourhood.

* Closeness Centrality is also quite similar, but "Network Science" seem to be closest to its neigbourhood.
