  * `rare.txt` ... ingredients/stopwords occuring only 1x in the `result.txt` - I excluded those
  * `stopwords.txt` ... manually crated list of stopwords from `result.txt`
  * `result-filtered.txt` ... file I use for analysis ( `result.txt` filtered with `rare.txt` and `stopwords.txt`)


# This is just some methods I use for analysis....

```python
import networkx as nx
from networkx.algorithms import community
import matplotlib.pyplot as plt
%matplotlib inline

def prepare_graph(kind):
    # Read the file...
    print("reading file... ", end="")
    recipes, ingredients, edges = set(), set(), []
    with open('result-filtered.txt', 'r') as f:
        for line in f:
            line = line.strip()
            recipe = line.split(",")[0]
            recipe_ingredients = line.split(",")[1:]
            recipes.add(recipe)
            ingredients = ingredients.union(set(recipe_ingredients))
            edges.extend([(recipe, i) for i in recipe_ingredients if (recipe, i) not in edges])
    recipes, ingredients = list(recipes), list(ingredients)
    print("recipes: {}, ingredients: {}".format(len(recipes), len(ingredients)))
    # Create graph...

    B = nx.Graph()
    # Add recipes
    B.add_nodes_from(recipes, bipartite=0)
    # Add ingredients
    B.add_nodes_from(ingredients, bipartite=1)
    # Add edges only between nodes of opposite node sets
    B.add_edges_from(edges)

    if kind == "recipes":
        X = recipes
    else:
        X = ingredients
    # A. Relations between recipes - 2 recipes are connected iff # common ingredient word >=1
    new_edges = []
    for idx_a in range(len(X)):
        for idx_b in range(idx_a+1, len(X)):
            a = X[idx_a]
            b = X[idx_b]
            r = sorted(nx.common_neighbors(B, a, b))
            if len(r) >= 1:
                new_edges.append((a,b))
    C = nx.Graph()
    C.add_nodes_from(X)
    C.add_edges_from(new_edges)
    # Communities correspond to recipes that use common ingredients?
    communities = sorted(community.greedy_modularity_communities(C), key=len, reverse=True)
    C_pos = nx.spring_layout(C, k=0.1)

    return B, C, C_pos, communities
```

```python
def print_info_about_communities(C, B, communities, names, cnt_common=10):

    print("-----------------------------------------------")
    print("{} communities: {}".format(names[0], len(communities)))
    print("-----------------------------------------------")
    for idx, c in enumerate(communities):
        print("Community [{}]:, size: {}, most central {}:".format(idx, len(c), names[0]))
        C_sub = C.subgraph(list(c))
        b = nx.betweenness_centrality(C_sub, weight=None, normalized=True)
        sorted_keys = sorted(b, key=b.get, reverse=True)
        for i in range(min(10, len(b))):
            print("\t{}".format(sorted_keys[i]))
    print("-----------------------------------------------")

    # Print most 10 popular ingredients from each Recipe community
    #  ... for each recipe - get its ingredients (neighbours)
    #  ... for each neighbour count how many recipes in this community...
    class CountDict:
        def __init__(self):
            self._dict = {}

        def insert(self, key):
            if key in self._dict:
                self._dict[key] += 1
            else:
                self._dict[key] = 1

        def get_sorted(self):
            return {k: v for k, v in sorted(self._dict.items(), key=lambda item: item[1], reverse=True)}

    populars = []
    for c in communities:
        count_dict = CountDict()
        for recipe in c:
            ingredients = nx.neighbors(B, recipe)
            for i in ingredients:
                if i not in count_dict._dict:
                    recipes = nx.neighbors(B, i)
                    for rr in recipes:
                        if rr in c:
                            count_dict.insert(i)
        res = count_dict.get_sorted()
        populars.append(res)

    for idx, popular in enumerate(populars):
        print("Community [{}]:".format(idx), "frequency of popular {}:".format(names[1]) )
        i = 0
        for k, v in popular.items():
            print("\t{} {}".format(k.ljust(10), v/len(communities[idx])))
            i+=1
            if i == cnt_common:
                break
```

```python
def plot_info_about_communities(C, communities, pos):
    def set_node_community(G, communities):
        for c, v_c in enumerate(communities):
            for v in v_c:
                # Add 1 to save 0 for external edges
                G.nodes[v]['community'] = c + 1

    def set_edge_community(G):
        for v, w, in G.edges:
            if G.nodes[v]['community'] == G.nodes[w]['community']:
                # Internal edge, mark with community
                G.edges[v, w]['community'] = G.nodes[v]['community']
            else:
                # External edge, mark as 0
                G.edges[v, w]['community'] = 0

    def get_color(i, r_off=1, g_off=1, b_off=1):
        r0, g0, b0 = 0, 0, 0
        n = 16
        low, high = 0.1, 0.9
        span = high - low
        r = low + span * (((i + r_off) * 3) % n) / (n - 1)
        g = low + span * (((i + g_off) * 5) % n) / (n - 1)
        b = low + span * (((i + b_off) * 7) % n) / (n - 1)
        return r, g, b

    plt.rcParams.update(plt.rcParamsDefault)
    plt.rcParams.update({'figure.figsize': (15, 10)})
    plt.style.use('dark_background')

    # Set node and edge communities
    set_node_community(C, communities)
    set_edge_community(C)

    # Set community color for internal edges
    external = [(v, w) for v, w in C.edges if C.edges[v, w]['community'] == 0]
    internal = [(v, w) for v, w in C.edges if C.edges[v, w]['community'] > 0]
    internal_color = ["black" for e in internal]
    node_color = [get_color(C.nodes[v]['community']) for v in C.nodes]
    # external
    nx.draw_networkx(
        C,
        pos=pos,
        node_size=0,
        edgelist=external,
        edge_color="silver",
        node_color=node_color,
        alpha=0.2,
        with_labels=False)
    # internal
    nx.draw_networkx(
        C, pos=pos,
        edgelist=internal,
        edge_color=internal_color,
        node_color=node_color,
        alpha=0.05,
        with_labels=False)
    plt.show()
```

# Analysis of Recipe comunities

### Here is the summary of the result:
* We have 2 comunities
* **Community [0]** represent meals which are not sweet
* **Community [1]** represent sweet meals, deserts and sweet courses in general
* Below you can see most central recipes of both communities
* I also dispaly which ingredients occurs the most in each category
   * so for instance we can see that about 80% of recipes from **Community [0]**  contains pepper and salt
   * on the other hand, most popular ingerediences in recipes from **Community [1]** are sugar, salt and butter
* Finally I also include graph with those 2 communities and connections
  * BLUE - **Community [0]**
  * GREEN -  **Community [1]**

```python
# Analyze Recipe Communities
B_rec, C_rec, C_pos_rec, C_communities_rec = prepare_graph(kind="recipes")

# Print info
print_info_about_communities(C_rec, B_rec, C_communities_rec, cnt_common=10, names=["Recipes", "ingredients"])

# Plot graph
plot_info_about_communities(C_rec, C_communities_rec, C_pos_rec)
```
